Given a non-negative integer `x`, return *the square root of* `x` *rounded down to the nearest integer*. The returned integer should be **non-negative** as well.

You **must not use** any built-in exponent function or operator.

- For example, do not use `pow(x, 0.5)` in c++ or `x ** 0.5` in python.

**Example 1:**
Input: x = 4
Output: 2
Explanation: The square root of 4 is 2, so we return 2.
```
Input: x = 4
Output: 2
Explanation: The square root of 4 is 2, so we return 2.

```

**Example 2:**

Input: x = 8
Output: 2
Explanation: The square root of 8 is 2.82842..., and since we round it down to the nearest integer, 2 is returned.
</aside>

````python
class solution1:
    def __init__(self):
        pass
    def approx_root(self,n)-> int:
        '''
        Description : This function will take the input of number n abd returns its  square root to 
        nearest integer
        :type n: int
        :rtype: int
               
        '''
        try:
            if(n<=1):
                return n
            if(n//2 ==1):
                return 1
            if(n//2 >1):
                l=1
                r =n//2
                while(l<r):
                    if(l*l == n):
                        return l
                    if(r**r == n ):
                        return r
                    mid = (l+r)//2
                    
                    if(mid*mid == n):
                        return mid
                    if(mid*mid < n):
                        if(((mid+1)*(mid+1)) >n):
                            return mid
                        else:
                            l = mid + 1
                    if(mid*mid > n): 
                        if(((mid-1)*(mid-1)) <n):
                            return mid-1
                        else:
                            r = mid-1
                    
            return l
                        
                    
            
            
        except Exception as e:
            print("error has occure ",e)

sol = solution1()
n=37
print(sol.approx_root(n))
````

A peak element is an element that is strictly greater than its neighbors.

Given a **0-indexed** integer array `nums`, find a peak element, and return its index. If the array contains multiple peaks, return the index to **any of the peaks**.

You may imagine that `nums[-1] = nums[n] = -∞`. In other words, an element is always considered to be strictly greater than a neighbor that is outside the array.

You must write an algorithm that runs in `O(log n)` time.

**Example 1:**

```
Input: nums = [1,2,3,1]
Output: 2
Explanation: 3 is a peak element and your function should return the index number 2.
```

**Example 2:**
Input: nums = [1,2,1,3,5,6,4]
Output: 5
Explanation: Your function can return either index number 1 where the peak element is 2, or index number 5 where the peak element is 6.
    
</aside>

````python
class solution2:
    def __init__(self):
        pass
    def peak_ele(self,nums)->int:
        '''
            Description: this function will take the input of list of elements and return the index of ele 
            which is greater than it's neighbour
            :type nums :list[int]
            :rtype : int
        
        '''
        try:
            l=1
            r= len(nums)-2
            while(l<r):
                if(nums[l] > nums[l-1] and nums[l] > nums[l+1]):
                    return l
                if(nums[r] > nums[r-1] and nums[r] > nums[r+1]):
                    return r 
                mid = l+r //2
                if(nums[mid] > nums[mid-1] and nums[mid] > nums[mid+1]):
                    return mid
                if(nums[mid] > nums[l]):
                    l=mid +1
                if(nums[mid] < nums[r]):
                    r = mid -1
            
        except Exception as e:
            print("error has occure ",e)

sol = solution2()  
nums = [1,2,1,3,5,6,4]
print(sol.peak_ele(nums))
````

````python
class solution3:
    def __init__(self):
        pass
    def missing_nums(self,nums)->int:
        '''
            Description: This function will take input nums and return the missing no from that list of n nums
            :type nums : list[int]
            :rtype : int
        
        '''
        try:
            nums.sort()
            l=0
            r = len(nums)-1
            while(l<r):
                
                mid = (l+r) //2
                if(nums[mid] == mid):
                    l = mid+1
                else:
                    r = mid-1
            return l
            
        except Exception as e:
            print("error has occure ",e)

sol = solution3()
nums = [9,6,4,2,8,5,7,0,1]
print(sol.missing_nums(nums))
````

Given an array of integers `nums` containing `n + 1` integers where each integer is in the range `[1, n]` inclusive.

There is only **one repeated number** in `nums`, return *this repeated number*.

You must solve the problem **without** modifying the array `nums` and uses only constant extra space.

**Example 1:**

```
Input: nums = [1,3,4,2,2]
Output: 2

```

**Example 2:**
Input: nums = [3,1,3,4,2]
Output: 3
</aside>

````python
class solution4:
    def __init__(self):
        pass
    def repeated_number(self,nums):
        '''
        Description : This functon will take list of numbers n and return the extra number which is double
        :type nums : list[int]
        :rtype : int
        '''
        try:
            for i in range(len(nums)):
                
                if(nums[i] > 0):
                    val = nums[i]-1
                    nums[val] = nums[val] * -1
                else:
                   
                    return (nums[i]*-1)
                
        except Exception as e:
            print("error has occure ",e)

sol = solution4()
nums = [3,1,1,4,2]
print(sol.repeated_number(nums))
````

Given two integer arrays `nums1` and `nums2`, return *an array of their intersection*. Each element in the result must be **unique** and you may return the result in **any order**.

**Example 1:**

```
Input: nums1 = [1,2,2,1], nums2 = [2,2]
Output: [2]

```

**Example 2:**
    
Input: nums1 = [4,9,5], nums2 = [9,4,9,8,4]
Output: [9,4]
Explanation: [4,9] is also accepted.

</aside>

````python
class solution5:
    def __init__(self):
        pass
    def intersection_num(self,n1,n2):
        '''
        Description : This function will take the input array nums and return the list of common integers
        :type n1: list[int]
        :type n2: list[int]
        :rtype : list
          
        '''
        try:
            return_list =[]
            for val in n1:
                if(val in n2):
                    return_list.append(val)
            return list(set(return_list))
            
        except Exception as e:
            print("error has occure ",e)

sol = solution5()
nums1 = [1,2,2,1]
nums2 = [2,2,1]
print(*sol.intersection_num(nums1,nums2))
````

Suppose an array of length `n` sorted in ascending order is **rotated** between `1` and `n` times. For example, the array `nums = [0,1,2,4,5,6,7]` might become:

- `[4,5,6,7,0,1,2]` if it was rotated `4` times.
- `[0,1,2,4,5,6,7]` if it was rotated `7` times.

Notice that **rotating** an array `[a[0], a[1], a[2], ..., a[n-1]]` 1 time results in the array `[a[n-1], a[0], a[1], a[2], ..., a[n-2]]`.

Given the sorted rotated array `nums` of **unique** elements, return *the minimum element of this array*.

You must write an algorithm that runs in `O(log n) time.`

**Example 1:**

```
Input: nums = [3,4,5,1,2]
Output: 1
Explanation: The original array was [1,2,3,4,5] rotated 3 times.

```

**Example 2:**

```
Input: nums = [4,5,6,7,0,1,2]
Output: 0
Explanation: The original array was [0,1,2,4,5,6,7] and it was rotated 4 times.

```

**Example 3:**
Input: nums = [11,13,15,17]
Output: 11
Explanation: The original array was [11,13,15,17] and it was rotated 4 times.
    
</aside>

````python
class solution6:
    def __init__(self):
        pass
    def min_rotated_array(self,nums):
        '''
           Description : This function will take the sorted rotated array and return the min number
           :type nums : list[int]
           :rtype : int
           
        '''
        try:
            if(nums[0] < nums[-1]):
                return nums[0]
            l,r=0,len(nums)-1
            while(l<r):
                mid = (l+r)//2
                
                if(l!=0):
                    if(nums[l] < nums[l+1] and nums[l] < nums[l-1]):
                        return nums[l]
                if(r!= len(nums)-1):
                    if(nums[r] < nums[r+1] and nums[r] < nums[r-1]):
                        return nums[r]
                    
                if(nums[mid] < nums[mid+1] and nums[mid] < nums[mid-1]):
                    return nums[mid]
                elif(nums[l] < nums[mid]):
                    l= mid+1
                elif(nums[mid] < nums[r]):
                    r = mid-1
                    
            
        except Exception as e:
            print("error has occure ",e)

sol = solution6()
nums = [3,4,5,1,2]
print(sol.min_rotated_array(nums ))
````

 **Question 8**

Given two integer arrays `nums1` and `nums2`, return *an array of their intersection*. Each element in the result must appear as many times as it shows in both arrays and you may return the result in **any order**.

**Example 1:**

```
Input: nums1 = [1,2,2,1], nums2 = [2,2]
Output: [2,2]

```

**Example 2:**
Input: nums1 = [4,9,5], nums2 = [9,4,9,8,4]
Output: [4,9]
Explanation: [9,4] is also accepted.
    
</aside>

````python
from collections import Counter 
class solution8:
    def __init__(self):
        pass
    def Inter_lists(self,s1,s2)-> list:
        '''
           Description : This function will take the input of two list of numbers and return the intersection of both
           :type s1 : list[int]
           :type s2 : list[int]
           :rtype : list
        '''
        try:
            counter1 = Counter(nums1)
            counter2 = Counter(nums2)
            return_list = []
            inter_count = (counter1 & counter2)
            for items in inter_count.items():
                for i in range(items[1]):
                    return_list.append(items[0])
            
            return return_list
            
        except Exception as e:
            print("error has occure ",e)

sol = solution8()
nums1 = [1,2,2,1]
nums2 = [2,2]
print(sol.Inter_lists(nums1,nums2))
````
