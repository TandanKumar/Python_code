# **Question 1**

Convert 1D Array Into 2D Array

You are given a **0-indexed** 1-dimensional (1D) integer array original, and two integers, m and n. You are tasked with creating a 2-dimensional (2D) array with Â m rows and n columns using **all** the elements from original.

The elements from indices 0 to n - 1 (**inclusive**) of original should form the first row of the constructed 2D array, the elements from indices n to 2 * n - 1 (**inclusive**) should form the second row of the constructed 2D array, and so on.

Return *an* m x n *2D array constructed according to the above procedure, or an empty 2D array if it is impossible*.

**Example 1:**

![img](sc.png)
**Input:** original = [1,2,3,4], m = 2, n = 2

**Output:** [[1,2],[3,4]]

**Explanation:** The constructed 2D array should contain 2 rows and 2 columns.

The first group of n=2 elements in original, [1,2], becomes the first row in the constructed 2D array.

The second group of n=2 elements in original, [3,4], becomes the second row in the constructed 2D array.

</aside>

# 

```python
class solution1:
    def __init__(self):
        pass
    def array_conversion(self,arr,m,n):
        '''
            This function will take the input an array of numbers,m,n. It will convert the array into m*N matrix and return 
            :type arr:list[int]
            :type m : int
            :type n :int
            :rtype : list[list[int]]
        '''
        try:
            row=0
            col = 0
            lis_conver = [[0 for i in range(n)] for j in range(m)]
            for i in range(len(arr)):
                if(col >= n):
                    row+=1
                    col=0
                lis_conver[row][col] = arr[i]
                col +=1
            return lis_conver
                
            
        except Exception as e:
            print("error has occure ",e)

sol = solution1()  
original = [1,2,3,4]
m = 2
n = 2
print(sol.array_conversion(original,m,n))
```

 **Question 3**
Given an integer array nums sorted in **non-decreasing** order, return *an array of **the squares of each number** sorted in non-decreasing order*.

**Example 1:**

Input: nums = [-4,-1,0,3,10]

Output: [0,1,9,16,100]

**Explanation:** After squaring, the array becomes [16,1,0,9,100].
After sorting, it becomes [0,1,9,16,100]

</aside>

```python
class solution3:
    def __init__(self):
        pass
    def square_function(self,arr):
        '''
            This function will take the input of array sorted in non-decreasing order.
            It will square the number and return it in sorted way
            :type arr : list[int]
            :type :int
        '''
        try:
            return sorted([val**2 for val in arr])
        except Exception as e:
            print("error has occure ",e)

sol = solution3()  
nums = [-4,-1,0,3,10]
print(sol.square_function(nums))
```

For nums2, nums2[0] = 2 is present at index 1 of nums1, whereas nums2[1] = 4 and nums2[2] = 6 are not present in nums2. Therefore, answer[1] = [4,6].

```python
class solution4:
    def __init__(self):
        pass
    def list_o_un_ele(self,arr1,arr2):
        '''
            This function will take input two arrays and retrun two dimensional list with uncommon element
            :type arr1 : list[int]
            :type arr2 : list[int]
            :rtype arr3 : list[[int],[int]]
        '''
        try:
            i,j=0,0
            lis = [[],[]]
            while(i < len(arr1)-1 and j< len(arr2)-1):
                if(arr1[i] == arr2[j]):
                    i+=1
                    j+=1
                elif(arr1[i] > arr2[j]):
                    if(arr2[j] not in lis[1]):
                        lis[1].append(arr2[j])
                    j+=1
                else:
                    if(arr1[i] not in lis[0]):
                        lis[0].append(arr1[i])
                    i+=1
            
            if(i < len(arr1)):
                while(i < len(arr1)):
                    lis[0].append(arr1[i])
                    i+=1
            if(j < len(arr2)):
                while(j < len(arr2)):
                    lis[1].append(arr2[j])
                    j+=1
            return lis
                
            
        except Exception as e:
            print("error has occure ",e)

sol = solution4()
nums1 = [1,2,3]
nums2 = [2,4,6]
print(sol.list_o_un_ele(nums1,nums2))
```

# **Question 5**

Given two integer arrays arr1 and arr2, and the integer d, *return the distance value between the two arrays*.

The distance value is defined as the number of elements arr1[i] such that there is not any element arr2[j] where |arr1[i]-arr2[j]| <= d.

**Example 1:**

**Input:** arr1 = [4,5,8], arr2 = [10,9,1,8], d = 2

**Output:** 2

**Explanation:**

For arr1[0]=4 we have:

|4-10|=6 > d=2

|4-9|=5 > d=2

|4-1|=3 > d=2

|4-8|=4 > d=2

For arr1[1]=5 we have:

|5-10|=5 > d=2

|5-9|=4 > d=2

|5-1|=4 > d=2

|5-8|=3 > d=2

For arr1[2]=8 we have:

**|8-10|=2 <= d=2**

**|8-9|=1 <= d=2**

|8-1|=7 > d=2

**|8-8|=0 <= d=2**

</aside>

```python
class solution5:
    def __init__(self):
        pass
    def dis_arr(self,arr1,arr2,d):
        '''
            This function will take the input two arrays of nums and distance and return the distance of arr1
            from arr2 where it is less than d
            :type arr1 : list[int]
            :type arr2 : list[int]
            :type d : int
            :rtype : int 
        
        '''
        try:
            arr2 = sorted(arr2)
            distance = 0
            for ele in arr1:
                start_index =0
                end_index = len(arr2)-1
                
                while(start_index < end_index):
                    mid_index = (start_index + end_index)//2
                    
                    if(abs(ele - arr2[start_index]) <=d):
                        distance +=1
                        start_index +=1
                    elif(abs(ele - arr2[end_index]) <=d):
                        distance +=1
                        end_index -=1
                    
                    if(abs(ele - arr2[mid_index]) <=d):
                      
                        distance +=1
                        break
                    elif(arr2[mid_index] < ele):
                        start_index +=1
                    elif(arr2[mid_index] >ele):
                        end_index -=1
            return distance
        except Exception as e:
            print("error has occure ",e)

sol = solution5()
arr1 = [4,5,8]
arr2 = [10,9,1,8]
d = 2
print(sol.dis_arr(arr1,arr2,d))
```

Given an integer array nums of length n where all the integers of nums are in the range [1, n] and each integer appears **once** or **twice**, return *an array of all the integers that appears **twice***.

You must write an algorithm that runs in O(n) time and uses only constant extra space.

**Example 1:**

**Input:** nums = [4,3,2,7,8,2,3,1]

**Output:**

[2,3]

</aside>

```python
class solution6:
    def __init__(self):
        pass
    def dupli(self,arr):
        '''
            This function will take the input of an array and return the non duplicates number
            :type arr :list[int]
            :rtype : int
        '''
        try:
            output =[]
            for num in nums:
                if nums[abs(num) - 1] < 0:
                    output.append(abs(num))
                else:
                    nums[abs(num) - 1] *= -1
            return output
        except Exception as e:
            print("error has occure ",e)

sol = solution6()  
nums = [4,3,2,7,8,2,3,1]
print(sol.dupli(nums))
```

An integer array original is transformed into a **doubled** array changed by appending **twice the value** of every element in original, and then randomly **shuffling** the resulting array.

Given an array changed, return original *if* changed *is a **doubled** array. If* changed *is not a **doubled** array, return an empty array. The elements in* original *may be returned in **any** order*.

**Example 1:**

**Input:** changed = [1,3,4,2,6,8]

**Output:** [1,3,4]

**Explanation:** One possible original array could be [1,3,4]:

- Twice the value of 1 is 1 * 2 = 2.
- Twice the value of 3 is 3 * 2 = 6.
- Twice the value of 4 is 4 * 2 = 8.

Other original arrays could be [4,3,1] or [3,1,4].

</aside>

```python
import collections

class solution8:
    def __init__(self):
        pass
    def double_array(self,arr):
        '''
            This function will take the input of a array which contains number and it's square.
            Return the original numbers
            :type arr : list[int]
            :rtype : list[int]
        
        '''
        try:
            if(len(arr)%2 !=0):
                return[]
                
            counter = collections.Counter(arr)
            res = []
            for ele in counter.keys():
                
                
                if(ele == 0):
                    if(counter[ele] %2 == 0):
                        res+=[ele]
                    else:
                        return []
                elif(ele > 0):
                    val = counter[ele]
                    if(val%2 ==0 and val//2 in counter):
                        val = val//2
                    
                    while val in counter:
                        if(counter[val] > 0):
                            res += [val] * counter[val] 
                            if counter[val+val] < counter[val]:
                                return []
                            counter[val+val] -= counter[val]
                            counter[val] = 0
                        val = val+1
                
            
            return res
        except Exception as e:
            print("error has occure ",e)

sol = solution8()
changed = [1,3,4,2,6,8]
print(sol.double_array(changed))
```
