 **Question 1**

A permutation perm of n + 1 integers of all the integers in the range [0, n] can be represented as a string s of length n where:

- s[i] == 'I' if perm[i] < perm[i + 1], and
- s[i] == 'D' if perm[i] > perm[i + 1].

Given a string s, reconstruct the permutation perm and return it. If there are multiple valid permutations perm, return **any of them**.

**Example 1:**

**Input:** s = "IDID"

**Output:**

[0,4,1,3,2]

</aside>

```python

class solution1:
    def __init__(self):
        pass
    def perm(self,S):
        '''
        This function will take the input of string in range[0,n].
        It will return the permutaion of the given number
        :type s: str
        :rtype: List[int]
        '''
        try:
            start, end = 0, len(S)
            return_value = []
            for x in S:
                if x == 'I':
                    return_value.append(start)
                    start += 1
                else:
                    return_value.append(end)
                    end -= 1

            return return_value + [start]
        except Exception as e:
            print("error has occure ",e)

sol = solution1()  
s = "IDID"
print(sol.perm(s))
```

 **Question 2**

You are given an m x n integer matrix matrix with the following two properties:

- Each row is sorted in non-decreasing order.
- The first integer of each row is greater than the last integer of the previous row.

Given an integer target, return true *if* target *is in* matrix *or* false *otherwise*.

You must write a solution in O(log(m * n)) time complexity.

**Example 1:**

![image](6_1.png)
**Input:** matrix = [[1,3,5,7],[10,11,16,20],[23,30,34,60]], target = 3

**Output:** true

</aside>

```python
class solution2:
    def __init__(self):
        pass
    def valid_integer(self,matrix,n):
        '''
            This function will take the input of matrix and number.
            If the number is present in the matrix, return true or false
            :type matrix : list[[int]]
            :type n : int
            contraint for matrix:
                    Each row is sorted in non-decreasing order.
                    The first integer of each row is greater than the last integer of the previous row.        
        '''
        try:
            for i in range(len(matrix)):
                if(matrix[i][-1] < n):
                    continue
                if(matrix[i][0] > n  ):
                    return False
                if(matrix[i][0] == n or matrix[i][-1]==n):
                        return True
                for j in range(0,len(matrix)-1):
                    if(matrix[i][j] == n):
                        return True
                    
            
        except Exception as e:
            print("error has occure ",e)

sol = solution2()  
matrix = [[1,3,5,7],[10,11,16,20],[23,30,34,60]]
target = 3
print(sol.valid_integer(matrix,target))
```

 **Question 3**

Given an array of integers arr, return *true if and only if it is a valid mountain array*.

Recall that arr is a mountain array if and only if:

- arr.length >= 3
- There exists some i with 0 < i < arr.length - 1 such that:
    - arr[0] < arr[1] < ... < arr[i - 1] < arr[i]
    - arr[i] > arr[i + 1] > ... > arr[arr.length - 1]

![image](6_2.png)
**Example 1:**

**Input:** arr = [2,1]

**Output:**

false

</aside>

```python
class solution3:
    def __init__(self):
        pass
    def mountain_array(self,array):
        '''
            This function will take input of an array and return boolean value wheter it is monotonic increasing or not
            :typr array : list[int]
            :rtype :boolean
        
        '''
        try:
            if(len(array) <=2):
                return False
            counter= 0
            for i in range(len(array)-1):
                if(array[i] > array[i+1]):
                    counter = i+1
                    break
            for i in range(counter,len(array)-2):
                if(array[i] < array[i+1]):
                    return False
            return True
        except Exception as e:
            print("error has occure ",e)

sol = solution3()
array =[0,2,3,4,5,2,1,0]

print(sol.mountain_array(array))
```

Given a binary array nums, return *the maximum length of a contiguous subarray with an equal number of* 0 *and* 1.

**Example 1:**

**Input:** nums = [0,1]

**Output:** 2

**Explanation:**

[0, 1] is the longest contiguous subarray with an equal number of 0 and 1.

</aside>

```python
class solution4:
    def __init__(self):
        pass
    def cont_sub_array(self,array):
        '''
             This function will take the input fo an array of number o and 1.
             It will return the longest subarray of 0 and 1
             :type array : list[int]
             :rtype : list[int]
        '''
        try:
            counter =0
            return_list =[]
            ls = []
            for ele in array:
                
                if(ele==0):
                    counter -=1
                    ls.append(0)
                if(ele==1):
                    counter +=1
                    ls.append(1)
                if(counter ==0):
                    if(len(return_list) < len(ls)):
                        return_list = ls
                    ls =[]
                    counter =0
            return len(return_list)
        except Exception as e:
            print("error has occure ",e)

sol = solution4()
array = [0,1]
print(sol.cont_sub_array(array))
```

 **Question 6**

An integer array original is transformed into a **doubled** array changed by appending **twice the value** of every element in original, and then randomly **shuffling** the resulting array.

Given an array changed, return original *if* changed *is a **doubled** array. If* changed *is not a **doubled** array, return an empty array. The elements in* original *may be returned in **any** order*.

**Example 1:**

**Input:** changed = [1,3,4,2,6,8]

**Output:** [1,3,4]

**Explanation:** One possible original array could be [1,3,4]:

- Twice the value of 1 is 1 * 2 = 2.
- Twice the value of 3 is 3 * 2 = 6.
- Twice the value of 4 is 4 * 2 = 8.

Other original arrays could be [4,3,1] or [3,1,4].

</aside>

```python

import collections

class solution6:
    def __init__(self):
        pass
    def double_array(self,arr):
        '''
            This function will take the input of a array which contains number and it's square.
            Return the original numbers
            :type arr : list[int]
            :rtype : list[int]
        
        '''
        try:
            if(len(arr)%2 !=0):
                return[]
                
            counter = collections.Counter(arr)
            res = []
            for ele in counter.keys():
                
                
                if(ele == 0):
                    if(counter[ele] %2 == 0):
                        res+=[ele]
                    else:
                        return []
                elif(ele > 0):
                    val = counter[ele]
                    if(val%2 ==0 and val//2 in counter):
                        val = val//2
                    
                    while val in counter:
                        if(counter[val] > 0):
                            res += [val] * counter[val] 
                            if counter[val+val] < counter[val]:
                                return []
                            counter[val+val] -= counter[val]
                            counter[val] = 0
                        val = val+1
                
            
            return res
        except Exception as e:
            print("error has occure ",e)

sol = solution6()
changed = [1,3,4,2,6,8]
print(sol.double_array(changed))
```

Given a positive integer n, generate an n x n matrix filled with elements from 1 to n2 in spiral order.

**Example 1:**

![image](6_3.png)
**Input:** n = 3

**Output:** [[1,2,3],[8,9,4],[7,6,5]]

</aside>

```python
class solution7:
    def __init__(self):
        pass
    def spiral_matrix(self,n):
        '''
        This function will take the input number from which it will construct n*n matrix
        :type n : int
        :type :list[int]
        
        '''
        try:
            return_list = [[0]*n for _ in range(n)]
            right = n
            left =0 
            top =0
            bottom=n
            counter =1

            while(counter < (n*n +1)):


                for i in range(left,right,1):
                    if(counter >n*n):
                        break
                    return_list[left][i] = counter
                    counter+=1

                top+=1
                for j in range(top,bottom,1):
                    if(counter > n*n):
                        break

                    return_list[j][right-1] = counter
                    counter +=1
                    if(counter >n*n):
                        break

                right-=1
                for k in range(right-1,left-1,-1):
                    if(counter > n*n):
                        break
                    return_list[bottom-1][k] = counter
                    counter +=1  
                bottom -=1

                for l in range(bottom -1,top-1,-1):
                    if(counter > n*n):
                        break

                    return_list[l][left] = counter
                    counter +=1
                left +=1
            return (return_list)
        except Exception as e:
            print("Error has occured",e)

sol = solution7()
n=4
print(sol.spiral_matrix(n))
```

Given two [sparse matrices](https://en.wikipedia.org/wiki/Sparse_matrix) mat1 of size m x k and mat2 of size k x n, return the result of mat1 x mat2. You may assume that multiplication is always possible.

**Example 1:**

![image](6_4.png)
**Input:** mat1 = [[1,0,0],[-1,0,3]], mat2 = [[7,0,0],[0,0,0],[0,0,1]]

**Output:**

[[7,0,0],[-7,0,3]]

</aside>

```python
class solution8:
    def __init__(self):
        pass
    def matrix_multi(self,arr1,arr2):
        '''
           This function will take the input two array and return the multiplication of the matrix
           :type arr1 :list[int]
           :type arr2 : list[int]
           :rtype  : list[int]
        '''
        try:
            return_matrix = [[0]* (len(arr2[0])) for _ in range(len(arr1))] 
            
            for i in range(len(arr1)):
                
                for j in range(len(arr2[0])):
                    
                    for k in range(len(arr2)):
                        return_matrix[i][j] += arr1[i][k] * arr2 [k][j]
                        
            return return_matrix
                        
        except Exception as e:
            print("error has occure ",e)

sol = solution8()
mat1 = [[1,0,0],[-1,0,3]]
mat2 = [[7,0,0],[0,0,0],[0,0,1]]
print(sol.matrix_multi(mat1,mat2))
```
