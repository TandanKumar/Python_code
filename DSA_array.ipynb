You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

**Example:**
Input: nums = [2,7,11,15], target = 9
Output0 [0,1]

**Explanation:** Because nums[0] + nums[1] == 9, we return [0, 1][

```python
import array as ar
import numpy as np
class solution1: # this class is solution for the first problem
    def __init__(self):
        pass
    def max_sum_brute(self, arr,tar): 
        '''
        This function will take two input  array and target value.
        This function will return indices from the array whose sum is equal to target.
        This function implement the brute force approach
        constraint : Two integers should not be same
        
        '''
        try :
            index_1 = 0 ## initializing the first index
            index_2=0   ## Initializing the second index
            if(len(arr) ==0): ## checking whetherarray is empty or not 
                return (0)
            ## Loop for finding the indexs for the target value
            for i in range(len(arr)): 
                for j in range(i+1,len(arr)):
                    temp = arr[i] + arr[j]
                    ## comparing with target with the sum with condition 
                    if((temp == tar) & (arr[i] != arr[j]) ): 
                        index_1 = i
                        index_2 = j
                        ## returning the array of indexes
                        return(ar.array('i',[index_1,index_2]))
                    else :
                        return (0)
        except EXCEPTION as e:
            print("Error has been occured  ",e)
            
    def max_sum_optm(self,arr,tar):
        '''
        This function will take two input  array and target value.
        This function will return indices from the array whose sum is equal to target.
        This function implement the optimum approach approach
        constraint : Two integers should not be same, given array is sorted
        
        '''
        try:
            
            if(len(arr) ==0): ## checking whetherarray is empty or not 
                return (0)
            index_1=0
            index_2 = len(arr)-1
            
            for i in range(len(arr)):
                if(((arr[index_1] + arr[index_2]) == tar)&  (arr[index_1] != arr[index_2])  ):
                    return(ar.array('i',[index_1,index_2]))
                elif((arr[index_1] + arr[index_2]) < tar):
                    index_1 = index_1 +1
                elif ((arr[index_1] + arr[index_2]) > tar):
                    index_2 = index_2  - 1
            return(0)
        except EXCEPTION as e:
            print("Error has occured ",e)
            
```

```python
sol = solution1()
arr = ar.array('i',[2,7,11,15])
tar = 9
index_brute = sol.max_sum_brute(arr,tar)
print("Index from brute",index_brute)
opt_sol = sol.max_sum_optm(arr,tar)
print("Index from the optimized solution",opt_sol)


```

 **Q2.** Given an integer array nums and an integer val, remove all occurrences of val in nums in-place. The order of the elements may be changed. Then return the number of elements in nums which are not equal to val.

Consider the number of elements in nums which are not equal to val be k, to get accepted, you need to do the following things:

- Change the array nums such that the first k elements of nums contain the elements which are not equal to val. The remaining elements of nums are not important as well as the size of nums.
- Return k.

**Example :**
Input: nums = [3,2,2,3], val = 3
Output: 2, nums = [2,2,_*,_*]

**Explanation:** Your function should return k = 2, with the first two elements of nums being 2. It does not matter what you leave beyond the returned k (hence they are underscores

</aside>

```python
import numpy as np
class solution2:
    def __init__(self):
        pass
    def occur_num(ar,val):
        '''
        Occur_num(ar,val) will take the input of array and integer.
        It will replace the occurance of val in array as '_', shifting in the last
        Return array of integers which are not replaced and replaced with '_' in the last.
        Also, count of non replced number in the array
        
        '''
        try:
            if(len(ar)==0): # checking whether array is null
                return(0,[0])
            start_index= 0
            last_index = len(ar)
            count = 0
            for i in range(0,last_index ):
                
                if((ar[start_index] == val) & (start_index != (last_index -1))):
                    
                    temp = ar[last_index-1]
                    ar[last_index-1] = '*'
                    ar[start_index]= temp
                    last_index = last_index -1
                    count = count +1
                    
                    
                elif((ar[start_index] == val)&(start_index == (last_index -1))):
                    count = count +1
                    ar[start_index] = '*' 
                    
                else:
                    start_index = start_index +1
                
            
            return(count,ar)
        except Exception as e:
            print("Error has occured",e)
                
```

```python
sol = solution2
arr = [3,2,2,3]
arr1 = [3,4,3,8,7,4,3,3,4,3,3,3,3]
val = 3
k,arr_o = sol.occur_num(arr,val)
print("{} nums : {}".format(k,arr_o))
k,arr_o = sol.occur_num(arr1,val)
print("{} nums : {}".format(k,arr_o))
```

 **Q3.** Given a sorted array of distinct integers and a target value, return the index if the target is found. If not, return the index where it would be if it were inserted in order.

You must write an algorithm with O(log n) runtime complexity.

**Example 1:**
Input: nums = [1,3,5,6], target = 5

Output: 2

</aside>

```python
class solution3:
    def __init__(self):
        pass
    def target_array_index(arr,tar):
        '''
        target_array_index(arr,tar) is a function which will take the input as array and target value.
        It will return the index of the target value in the array if it is found 
        Otherwise it will retrun the index where it should be placed
        
        '''
        try: 
            if(len(arr)==0): # checking whether arr is empty
                return 0
            elif(tar <= arr[0]): # checking whether first element of array is equal or less than target
                return 0
            elif(tar >= arr[-1]): # checking whether last element of array is equal or more than taget
                return (len(arr))
            first_index = 0
            last_index = len(arr)-1

            while (first_index < last_index): # loop for finding the target number in the array

                mid_index =  int((first_index +last_index)/2)

                if(arr[mid_index] == tar): # checking the middle value for target
                    return mid_index
                # shifting the last index as arr on mid_index is greater than target
                elif((arr[mid_index] > tar)&((last_index - first_index)!=1)): 

                    last_index = mid_index
                # shifting the first index as arr on mid_index is less than target    
                elif((arr[mid_index] < tar)&((last_index - first_index)!=1)):
                    first_index = mid_index
                # returning the lastindex    
                elif((last_index - first_index) == 1): 
                    return last_index
        except Exception as e:
            print("Error has occured",e)
            
            
        
```

```python
sol= solution3
arr = [1,3,5,6]
tar = 5
print(sol.target_array_index(arr,tar))
```

 **Q5.** You are given two integer arrays nums1 and nums2, sorted in non-decreasing order, and two integers m and n, representing the number of elements in nums1 and nums2 respectively.

Merge nums1 and nums2 into a single array sorted in non-decreasing order.

The final sorted array should not be returned by the function, but instead be stored inside the array nums1. To accommodate this, nums1 has a length of m + n, where the first m elements denote the elements that should be merged, and the last n elements are set to 0 and should be ignored. nums2 has a length of n.

**Example 1:**
Input: nums1 = [1,2,3,0,0,0], m = 3, nums2 = [2,5,6], n = 3
Output: [1,2,2,3,5,6]

**Explanation:** The arrays we are merging are [1,2,3] and [2,5,6].
The result of the merge is [1,2,2,3,5,6] with the underlined elements coming from nums1.

</aside>

```python
class solution5:
    
    def  __init__(self):
        pass
    def merge(self, arr1, m, arr2, n):
        '''
        merge function will take input two arrays with length with m and m+n
        arr1 will have numbers till m position and after that it will have 0
        arr2 will have size of n
        
        :type nums1: List[int]
        :type m: int
        :type nums2: List[int]
        :type n: int
        '''
        
        try:
            i= m-1
            j= n-1
            k= m+n-1
            while((i>=0 ) &(j>=0)):
                if(nums1[i] > nums2[j]):
                    nums1[k] =nums1[i]
                    k=k-1
                    i=i-1
                else:
                    nums1[k] =nums2[j]
                    k=k-1
                    j=j-1
            while(j>=0):
                nums1[k]= nums2[j]
                k = k-1
                j=j-1
            return nums1
        except Exception as e:
            print("Erro has occured ",e)
```

```python
nums1 = [1,2,3,0,0,0]
m = 3
nums2 = [2,5,6]
n = 3 
sol = solution5()
output_array = sol.merge(nums1,m,nums2,n)
print(output_array)
```

**Example 1:**
Input: nums = [1,2,3,1]

Q7. Given an integer array nums, move all 0's to the end of it while maintaining the relative order of the nonzero elements.
Note that you must do this in-place without making a copy of the array.
Example 1:
Input: nums = [0,1,0,3,12]
Output: [1,3,12,0,0]

```python
class solution7:
    def __init__(self):
        pass
    def function_with_end_zero(self,arr):
        '''
           This function will accept the array and return the array with all the zeros at the end
           constraint : 
                           1. No extra space  or other will be used
                           2. all the non zero number should be in the same order
           :type arr :list[int] 
        
        '''
        try:
            count = 0
            for i in range(len(arr)):
                if(arr[i]!=0):
                    arr[count] = arr[i]
                    count += 1
            while(count < len(arr)):
                arr[count] = 0
                count +=1
            return arr
        except Exception as e:
            print("error has occured ",e)

arr = [0,1,0,3,12]
sol = solution7()
print(sol.function_with_end_zero(arr))
            
```

8.You have a set of integers s, which originally contains all the numbers from 1 to n. Unfortunately, due to some error, one of the numbers in s got duplicated to another number in the set, which results in repetition of one number and loss of another number.
You are given an integer array nums representing the data status of this set after the error.
Find the number that occurs twice and the number that is missing and return them in the form of an array.
Example 1:
Input: nums = [1,2,2,4]
Output: [2,3]

```python
class solution8:
    def __init__(self):
        pass
    def missing_number(self,arr):
        '''
            This function will take the input list and will check which number is replced by other number.
            It will return the old_number and replced number
             :type arr :list[int] 
        
        '''
        try:
            set_arr = set(arr)
            n = len(arr)
            missing_value = 0
            for i in range(1,n+1):
                if (i not in set_arr):
                    missing_value = i
                    break
            sum_array= sum(arr)
            sum_n_number = (n*(n+1))//2
            repeating_number = missing_value - (sum_n_number - sum_array)
            return([repeating_number,missing_value])
        
        except Exception as e:
            print("Error has occured",e)
            
sol= solution8()
arr = [1,2,2,4]
print(sol.missing_number(arr))
```
